## 1.3.1 Categorical and Binary Data

I'll start by exploring the `Level` and `EventTemplate` columns — both of which are categorical.

I want to compute:
- Mode (most common level of event)
- Expected value approximation (useful in operational planning)
- Visual distribution (bar and pie charts)

Understanding this breakdown is useful in SRE to:
- Detect excessive error-level events
- Profile the most frequent operational logs


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

url = "https://raw.githubusercontent.com/logpai/loghub/refs/heads/master/Apache/Apache_2k.log_structured.csv"
df = pd.read_csv(url)
df['Time_parsed'] = pd.to_datetime(df['Time'], errors='coerce')
df['Hour'] = df['Time_parsed'].dt.hour
df['DayOfWeek'] = df['Time_parsed'].dt.day_name()
df.head()
```

## 1.3.2 Correlation Between Variables

### Exploring Correlation Between Numeric Variables

```python
numeric_cols = df.select_dtypes(include='number')[['Hour']]
numeric_cols['LineId'] = df['LineId']

corr_pearson = numeric_cols.corr(method='pearson')
corr_spearman = numeric_cols.corr(method='spearman')
corr_kendall = numeric_cols.corr(method='kendall')

print("Pearson Correlation:")
print(corr_pearson)
print("\nSpearman Correlation:")
print(corr_spearman)
print("\nKendall Correlation:")
print(corr_kendall)
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(corr_pearson, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pearson Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x="Hour", y="LineId", data=numeric_cols, alpha=0.3)
plt.title("Scatterplot: Hour vs Log Entry Order")
plt.xlabel("Hour of Day")
plt.ylabel("Log Line ID")
plt.tight_layout()
plt.show()
```

Even with a limited number of numeric variables, correlation analysis gives me a quick way to:

- Spot **time-related accumulation** (e.g., more logs later in the day)
- See whether log ordering aligns with expected hourly load
- Decide whether trends are linear (Pearson), monotonic (Spearman), or consistent in rank (Kendall)

This practice is essential when designing **SLO dashboards** or building **baselines for alerting** that depend on combined signals.

#### Interpreting the Correlation Results

The correlation between `Hour` and `LineId` gives me insight into the log accumulation trend throughout the day.

- **Pearson (0.43)** indicates a moderate linear relationship: the later the hour, the higher the typical log line ID.
- **Spearman (0.46)** and **Kendall (0.43)** confirm that this relationship is **monotonic**, even if not strictly linear.

##### What does this mean?

As the day progresses, I observe a natural increase in the number of logs. This could reflect:
- More user or system activity later in the day
- Batch or background processes that kick in after a certain hour
- Normal time-driven accumulation of events

##### Why it matters in SRE and observability:

- Helps distinguish **expected patterns** from anomalies (e.g., a spike at 3 AM may be truly abnormal)
- Supports setting **alert windows** that adapt to hourly baseline
- Encourages time-aware observability — not all hours behave the same, and my alerts shouldn’t either

#### Analyzing the `Level` column

```python
level_counts = df['Level'].value_counts()
level_counts
```

```python
mode_level = df['Level'].mode()[0]
print(f"The most common log level is: {mode_level}")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Level', data=df, order=level_counts.index, palette='Set2')
plt.title("Log Level Distribution")
plt.ylabel("Count")
plt.xlabel("Log Level")
plt.show()
```

```python
plt.figure(figsize=(5, 5))
df['Level'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('Set2'), startangle=90)
plt.title("Log Level Proportion")
plt.ylabel("")
plt.show()
```

These charts help me visually understand the dominance of certain log types.

For example, if `notice` is overwhelmingly common, it may drown out the visibility of `error` logs — which would be a critical concern in production monitoring.

As someone working with observability, I always look for **imbalance in event levels**, as it can:
- Distort alert thresholds
- Obscure signals from failures
- Reveal overly verbose or under-instrumented components

#### Analyzing the `EventTemplate` column

```python
template_counts = df['EventTemplate'].value_counts().head(10)
template_counts
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(y=template_counts.index, x=template_counts.values, palette='muted')
plt.title("Top 10 Most Frequent Event Templates")
plt.xlabel("Count")
plt.ylabel("Event Template")
plt.tight_layout()
plt.show()
```

This view allows me to quickly identify what kinds of messages dominate the logs.

In production environments, I often want to know:

- Are critical events underrepresented?
- Could template consolidation reduce noise?

Understanding the **distribution of event types** helps prioritize monitoring focus and detect logging bloat.

## 1.3.3 Exploring Relationships Between Multiple Variables

In this section, I want to explore how multiple variables interact. This is useful when trying to:

- Detect compound patterns that aren't visible when looking at variables in isolation
- Identify hotspots, correlations or behavioral clusters
- Build better mental models for how logs reflect system dynamics

I'll use tools like:
- **Contingency tables** (e.g., `Level` vs `EventId`)
- **Hexbin plots** (volume by hour)
- **Contour plots**
- **Violin plots**

```python
top_ids = df['EventId'].value_counts().nlargest(10).index
contingency = pd.crosstab(df[df['EventId'].isin(top_ids)]['Level'],
                          df[df['EventId'].isin(top_ids)]['EventId'])
contingency
```

```python
plt.figure(figsize=(8, 5))
plt.hexbin(df['Hour'], df['LineId'], gridsize=25, cmap='YlOrRd', mincnt=1)
plt.colorbar(label="Log Density")
plt.title("Hexbin Plot: Hour vs Log Line ID")
plt.xlabel("Hour")
plt.ylabel("Log Order (LineId)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.violinplot(x='Hour', y='LineId', data=df, scale='width', inner='quartile', palette="Set3")
plt.title("Violin Plot of Log Line ID by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Log Order")
plt.tight_layout()
plt.show()
```
