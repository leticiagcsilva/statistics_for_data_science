```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm, probplot

sns.set(style="whitegrid")
```

```python
url = "https://raw.githubusercontent.com/logpai/loghub/refs/heads/master/Apache/Apache_2k.log_structured.csv"
df = pd.read_csv(url)
df['Time_parsed'] = pd.to_datetime(df['Time'], errors='coerce')
df['Hour'] = df['Time_parsed'].dt.hour
```

### 1.4.1 Empirical distributions and symmetry

### Note on Visualization Reuse

Although I’ve previously used histogram and boxplot visualizations in Day 2 to examine statistical variability, I now return to them with a different focus.

In this section, I’m not interested in quantifying dispersion (e.g., standard deviation, IQR) as much as I am in understanding the **shape** of the distribution:

- Is it symmetric or skewed?
- Does it have a long tail?
- Would a Gaussian (normal) assumption hold here?

These are the kinds of questions that impact statistical modeling, threshold setting, and alert design — especially in real-world observability contexts.


```python
hourly_counts = df['Hour'].value_counts().sort_index()

hourly_counts
```

**Visualizing the distribution of logs by hour**

In this step I want to observe how the frequency of logs behaves throughout the hours of the day.

```python
plt.figure(figsize=(10, 5))
sns.histplot(hourly_counts, bins=12, kde=True, color="steelblue")
plt.title("Hourly Log Count Distribution")
plt.xlabel("Log Counts")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 2))
sns.boxplot(x=hourly_counts, color="salmon")
plt.title("Boxplot of Hourly Log Counts")
plt.xlabel("Log Counts")
plt.tight_layout()
plt.show()
```

So far, I've noticed that the distribution of the number of logs per hour is not at all balanced. Most of the time, the system generates few events — but at some specific times, there is an explosion of logs.

This asymmetry was clear in both the histogram and the boxplot, which showed a long tail pulling the values ​​to the right. This is the kind of behavior that usually goes unnoticed when we only look at the average.

In practice, this shows me that using the average as a reference can be dangerous. The system seems "normal" most of the time, but there are windows of behavior that are very non-standard — and it is precisely these moments that interest me the most when I think about reliability and alerts.

### 1.4.2 Comparing with theoretical distributions

**Comparing with theoretical distributions**

Now I want to understand how far the distribution of the hourly log count is from a normal (Gaussian) distribution.

```python
plt.figure(figsize=(6, 6))
probplot(hourly_counts, dist="norm", plot=plt)
plt.title("QQ Plot - Comparison with Normal")
plt.xlabel("Theoretical quantiles")
plt.ylabel("Sample quantiles")
plt.tight_layout()
plt.show()
```

The QQ plot shows me that the real distribution has a heavier right tail (high values ​​that do not align with the normal), reinforcing the hypothesis of asymmetry and unexpected behavior.

This visualization already anticipates that **metrics such as mean and standard deviation can be misleading**.

### 1.4.3 Normality Tests

To complement the visual analysis, I want to apply formal statistical tests to check whether the log volume distribution follows a normal distribution.

I'll use:
- **Shapiro-Wilk Test** (good for small to medium samples)
- **D’Agostino and Pearson’s Test** (assesses skewness and kurtosis jointly)


```python
from scipy.stats import shapiro, normaltest

# Shapiro-Wilk test
shapiro_stat, shapiro_p = shapiro(hourly_counts)
print(f"Shapiro-Wilk Test: statistic={shapiro_stat:.4f}, p-value={shapiro_p:.4f}")

# D’Agostino and Pearson test
dagostino_stat, dagostino_p = normaltest(hourly_counts)
print(f"D’Agostino-Pearson Test: statistic={dagostino_stat:.4f}, p-value={dagostino_p:.4f}")
```

Both tests return **p-values well below 0.05**, meaning I have enough evidence to reject the null hypothesis that the data is normally distributed.

This confirms what I observed visually in the histogram and QQ plot — the log volume distribution is clearly **not Gaussian**, and I need to be cautious with statistical methods that assume normality.


### 1.4.4 Practical Implications for Observability

When designing thresholds, alerts, and SLOs, it's common to rely on average behavior. However, in this dataset the average log volume is misleading due to the skewed distribution and heavy right tail.

Instead of using the mean, it's more appropriate to use robust estimators such as:
- The **median**, which is resistant to extreme values
- The **p95**, to identify typical peaks without overreacting to outliers
- The **IQR**, to model acceptable ranges of fluctuation

Assuming normality when it doesn’t hold can result in:
- Too many false alerts
- Thresholds that are either too loose or too tight
- Poor understanding of actual system volatility

Understanding the distribution shape of operational data is essential for building **trustworthy alerting mechanisms**.
